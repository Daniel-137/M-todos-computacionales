## OSCILADOR ARMÓNICO

```python
import numpy as np
import matplotlib.pyplot as plt
import pylab as py
import pandas as pd
%pylab inline
```

## Variables del sistema

```python
A=10 #Amplitud
k=20 #
m=2 #Masa
#w=50
w=(k/m)**0.5  #Frecuencia 
```

## Posición

```python
t=np.arange(0,10,0.01)
#t1=np.arange(1,5,0.01)
x = A*np.cos(w*t)
#print(x)
#print(t)
```

## Velocidad

```python
v = -A*w*np.sin(w*t)
#print(v)
```

## Aceleración

```python
a = -A*(w**2)*np.cos(w*t)
#print(a)
```

## Energía potencial

```python
Ep = 0.5*k*(x**2)
#print(Ep)
```

## Energía cinética

```python
Ek = 0.5*m*(v**2)
#print(Ek)
```

## Gráficas

```python
plt.plot(t,x)
plt.xlabel('t')
plt.ylabel('X(t)')
plt.title('Posición vs Tiempo')
plt.show()

plt.plot(t,v)
plt.xlabel('t')
plt.ylabel('V(t)')
plt.title('Velocidad vs Tiempo')
plt.show()

plt.plot(t,a)
plt.xlabel('t')
plt.ylabel('a(t)')
plt.title('aceleración vs Tiempo')
plt.show()


```

## Arreglos de numpy

```python
x=np.arange(0,11,1)
def f(x): #Definimos la función
    return np.sin(x)*np.log(1*x+1)*np.sinh(x)
y=f(x)
tabla=np.zeros((len(x),3))
for i in range(len(x)):
    tabla[i] = [i , x[i], y[i]]
print(tabla)

np.savetxt('arreglos1.txt', tabla, delimiter = ' ') #Guardamos en un archivo por el método .savetxt
tabla = str(tabla)
archivo = open("arreglos.txt","w") #Creamos un archivo
archivo.write(tabla) #Escribimos en el archivo 
archivo.close()

archivo = open("arreglos.txt","r")
print(archivo.read())
archivo.close()
```

## Expansión de taylor para la función coseno

```python
def factorial(x): #Función factorial
    if x==1 or x==0:
        return 1
    else:
        return x*factorial(x-1)
#print(factorial(3))
def cos(x,n): #Función coseno con aproximación por serie de taylor
    c=0
    for i in range(n):
        c = c + ((-1)**i)*(x**(2*i))/factorial(2*i)
    return c
print("Aproximación con 2 términos: ",cos(np.pi/2,2))
print("Aproximación con 5 términos: ",cos(np.pi/2,5))
print("Aproximación con 10 términos: ",cos(np.pi/2,10))

#Usando la función importada desde numpy

print("Valor arrojado por la función importada desde numpy: ",np.cos(np.pi/2))
```
